Πολλές φορές όταν θέλουμε να περιγράψουμε μία ακολουθία DNA η οποία έχει δημιουργηθεί από μεταλλάξεις μίας άλλης ακολουθίας χρησιμοποιούμε τη κωδικοποίηση CIGAR. Η κωδικοποίηση CIGAR αποτελείται από τα γράμμα M και D. M σημαίνει match και D σημαίνει DELETION. Σε κάθε M ή D υπάρχει ένα αριθμός ο οποίος δηλώνει πόσα matches/deletions έχει η καινούργια ακολουθία. Αν αυτός ο αριθμός είναι 1 τότε παραλείπεται. Για παράδειγμα έστω το CIGAR ```2MD3M2D2M```. Αν το εφαρμόσουμε στην ακολουθία ```AACGCTT``` τότε έχουμε:
```
A A _ C G C _ _ T T
A A   C G C     T T 
 2M D   3M   2D  2M

```

Άρα το αποτέλεσμα της εφαρμογής του CIGAR ```2MD3M2D2M``` στην ακολουθία ```AACGCTT``` είναι: ```AA_CGC__TT```. Εκεί που υπάρχει Deletion βάζουμε underscore: ```_```.

Φτιάξτε τη συνάρτηση  ```validate_cigar``` η οποία θα παίρνε ένα CIGAR string και θα επιστρέφει True/False ανάλογα με το αν είναι όντως cigar string. Για παράδειγμα:

```
validate_cigar('2MD3M2D2M') --> True
validate_cigar('Hello') --> False
validate_cigar('2M1D3M2D2M') --> False (το 1 δεν επιτρέπεται)
validate_cigat('10M5M') --> False (δεν μπορείς ένα M να ακολουθεί ένα άλλο M)
validate_cigat('10D5D') --> False (δεν μπορείς ένα D να ακολουθεί ένα άλλο D)
```

```
import random
def rand_sam(n):
    
    chromosomes = list(range(1,23)) + ['X', 'Y']
    fn = 'test_3_13_file.txt'
    
    with open('test_3_13_file.txt', 'w') as f:
        for i in range(n):
            random_chromosome = random.choice(chromosomes)
            random_position = random.randint(1,1000000)
            
            f.write('CHR{} TEST {}\n'.format(random_chromosome, random_position))
            
    print ('Created file ', fn)
            
rand_sam(100)
```
